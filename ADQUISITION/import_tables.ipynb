```python
#IMPORTAMOS EL ARCHIVO DE SQL QUE HEMOS GUARDADO COMO UN .JSON() LA PRIMERA SE LLAMA BICIMAD_STATIONS.JSON


dataframe_bicimadstations = pd.read_json("../DATA/bicimad_stations.json")
dataframe_bicimadstations

```

```python
#AHORA TENEMOS QUE SEPARAR LA LATITUD Y LONGITUD EN DOS COLUMNAS PARA PODER CRUZARLA CON LA TABLA QUE ME HA TOCADO:

dataframe_bicimadstations #esta es la tabla que tenemos en la que la ultima columna tenemos que separarla en dos.

dataframe_bicimadstations['LONGITUD'] = [float(index.split(',')[0].replace("[", "")) for index in dataframe_bicimadstations['geometry_coordinates']]
    
dataframe_bicimadstations['LATITUD'] = [float(index.split(',')[1].replace("]", "")) for index in dataframe_bicimadstations['geometry_coordinates']]

#YA HEMOS DIVIDIDO LA COLUMNA FINAL EN DOS COLUMNAS UNA LLAMADA LATITUD Y OTRA LLAMADA LONGITUD.

dataframe_bicimadstations  #AQUI TENEMOS LA COMPROBACION DE QUE LA ULTIMA COLUMNA LA HEMOS SEPARADO EN OTRAS DOS COLUMNAS.


    
```

```python
#VAMOS A CREAR UN DATAFRAME SOLO CON LAS COLUMNAS QUE VAMOS A NECESITAR PARA COMPARAR CON LA TABLA QUE ME HA TOCADO A MI:
from shapely.geometry import Point
import geopandas as gpd   # conda install -c conda-forge geopandas

def to_mercator(lat, long):
    # transform latitude/longitude data in degrees to pseudo-mercator coordinates in metres
    c = gpd.GeoSeries([Point(lat, long)], crs=4326)
    c = c.to_crs(3857)
    return c


dataframe_bicis = pd.DataFrame(dataframe_bicimadstations[["name","address","LATITUD","LONGITUD"]])
dataframe_bicis["DISTANCIA_BICIS"] = dataframe_bicis.apply(lambda x : to_mercator(x["LATITUD"],x["LONGITUD"]),axis=1)
dataframe_bicis

#CUANDO QUEREMOS PILLAR MAS DE UNA COLUMNA VAMOS A TENER QUE PONER DOS CORCHETES, MUCHO CUIDADO.
#TENEMOS QUE COMPROBAR QUE TENEMOS EL MISMO NUMERO DE FILAS QUE EL ORIGINAL YA QUE ESO NO LO HEMOS MODIFICADO
#LA COLUMNA DEL INDEX ME LA PILLA DIRECTAMENTE.
dataframe_bicis
```

```python
#VAMOS A HACER EL MERGE DE LAS DOS TABLAS PARA QUE ME SAQUE TODAS LAS POSIBILIDADES CON UN MERGE ENTRE LAS DOS TABLAS.
dataframe_total = pd.merge(dataframe_sedes,dataframe_bicis, how = "cross")  #PARA PODER JUNTAR LAS DOS TABLAS CON TODAS LAS OPCIONES POSIBLES.
#dataframe_total
dataframe_total["TIPO DE LUGAR"] = "SEDES. Centros con espacios deportivos"
dataframe_total["DISTANCIA TOTAL"] = 0 #dataframe_total.apply(lambda x: distance_meters(x["LATITUD_y"],x["LONGITUD_y"],x["LATITUD_x"],x["LONGITUD_x"]) )  #PARA CREAR ESA COLUMNA, COMO METEMOS MAS DE UN ARGUMENTO

dataframe_total2 = dataframe_total[["NOMBRE","NOMBRE-VIA","TIPO DE LUGAR","DISTRITO","DISTANCIA_SEDES","name","address","DISTANCIA_BICIS","DISTANCIA TOTAL"]]
dataframe_total2


```

```python
from shapely.geometry import Point
import geopandas as gpd 
def distance_meters(DISTANCIA_SEDES, DISTANCIA_BICIS):
    # return the distance in metres between to latitude/longitude pair point in degrees (i.e.: 40.392436 / -3.6994487)
    
    return DISTANCIA_SEDES.distance(DISTANCIA_BICIS)

dataframe_total2["DISTANCIA TOTAL"] =  dataframe_total2.apply(lambda x: distance_meters(x["DISTANCIA_SEDES"],x["DISTANCIA_BICIS"]),axis=1 )
dataframe_total2

```

```python

#ahora tenemos que hacer un input para que el usuario meta la direccion que el quiera, y me de la estacion de bicimad mas cercana a la direccion que el usuario meta.

def bicimad_mas_cercana():
    i = str(input("Introduzca lugar"))
    a = dataframe_total2[dataframe_total2["NOMBRE"] == i]
    return a.sort_values(by = "DISTANCIA TOTAL", ascending = True).head(1)
bicimad_mas_cercana()




```

```python
#AHORA QUEREMOS QUE NOS SAQUE TODAS LAS ESTACIONES PARA CADA UNA DE LAS DIRECCIONES, POR LO TANTO LO UNICO QUE TENEMOS QUE HACER ES QUITAR EL HEAD()


def todas_las_estaciones():
    i = str(input("Introduzca lugar"))
    a = dataframe_total2[dataframe_total2["NOMBRE"] == i]
    return a.sort_values(by = "DISTANCIA TOTAL", ascending = True)
todas_las_estaciones()

```
